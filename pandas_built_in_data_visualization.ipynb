```python
import numpy as np, pandas as pd
%matplotlib inline
```

```python
dates = pd.date_range('20120601', periods=500)
d1 = pd.DataFrame(np.random.randn(500,4),index=dates,columns=['a','b','c','d'])
d1.head()
```

```python
d1['a'].hist(bins=20)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
d1['a'].plot(kind='hist',bins=20)
#or d1['a'].plot.hist()
```

```python
d1.plot.line(x=d1.index,y='c',figsize=(15,3),lw=2)
```

```python
d1.plot.scatter(x='a',y='c',s=50,figsize=(15,3))
```

```python
d1.plot.scatter(x='a',y='c',c='d',figsize=(15,3),cmap='coolwarm')
```

```python
d1.plot.scatter(x='a',y='b',s=d1['c']*100,figsize=(15,3))
```

```python
d1.plot.box()
```

```python
d1.plot.hexbin(x='a',y='b',figsize=(7,7),gridsize=30)
```

```python
d1.plot.kde()
```
