```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#to display output of plotting commands inline
%matplotlib inline 
```

```python
x = np.linspace(1,7,13)
y = x**2
x
```

```python
#functional way of displaying plots
plt.plot(x,y)
plt.title('y = X^2')
plt.xlabel("X axis")
plt.ylabel("Y axis")
#plt.show()
```

```python
#creating multiple plots on the same canvas
#parameters (no_of_rows, no_of_columns, plot_no)
plt.subplot(1,3,1)
plt.plot(x,y,'r')
plt.subplot(1,3,2)
plt.plot(y,x,'g')
plt.subplot(1,3,3)
plt.plot(x,x,'b')
```

```python
#Object oriented way of displaying plots
fig = plt.figure()#canvas
#fig.add_axes([left,bottom,width,height])
axes1 = fig.add_axes([0.1,0.5,0.7,0.7])
axes2 = fig.add_axes([0.2,0.6,0.4,0.4])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.5,0.7,0.7])
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("y = x^2")
axes1.plot(x,y)

axes2 = fig.add_axes([0.2,0.8,0.3,0.3])
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("y = x")
axes2.plot(y,y)
```

```python
figures,axes = plt.subplots(nrows=2,ncols=3)

plt.tight_layout()#to automatically adjust overlaps between plots 
```

```python
axes
```

```python
#Subplots automaticallyt sets axes
#tuple unpacking
figures,axes = plt.subplots(nrows=2,ncols=3)
for i in axes.reshape(1*6):
    i.plot(x,y)
    
axes[0][0].plot(x,y)   
axes[0][0].set_title(" y = X^2")
plt.tight_layout()
    
```

```python
axes
```

```python
#figsize, aspect ratio and dpi(dots per inch)
figures,axes = plt.subplots(nrows=3,ncols=1,figsize=(7,5))
axes[0].plot(x,x**3,label="cube")
axes[1].plot(x,x**2,label="square")
axes[2].plot(x,x,label="linear")
axes[0].legend()
axes[1].legend(loc=5)#location code: 5
axes[2].legend(loc=(0.3,0.1))#coordinates
plt.tight_layout()
```

```python
#color, linewidth, alpha
figures,axes = plt.subplots(nrows=3,ncols=1,figsize=(7,5))
axes[0].plot(x,y,color='red')#color='#f234ff' rgb hex code,
axes[1].plot(x,y,color='#ff0123',linewidth=7)
axes[2].plot(x,y,color='#ff0123',lw=7,alpha=0.5)#alpha (for adding transparency) #lw=linewidth
```

```python
#linestyle
figures,axes = plt.subplots(nrows=4,ncols=1,figsize=(7,7))
axes[0].plot(x,y,color='purple',linestyle='--')
axes[1].plot(x,y,color='#ff0123',ls='-.')#ls = linestyle
axes[2].plot(x,y,color='#ff0123',linestyle=':')
axes[3].plot(x,y,color='green',linestyle='steps')
plt.tight_layout()
```

```python
#marking points
figures,axes = plt.subplots(nrows=6,ncols=1,figsize=(7,7))
axes[0].plot(x,y,color='purple',linestyle='--',marker='o')
axes[1].plot(x,y,color='#ff0123',ls='-.',marker='+')#ls = linestyle
axes[2].plot(x,y,color='#ff0123',ls=':',marker='*')
axes[3].plot(x,y,color='green',ls='-',marker='2',markersize=10)
axes[4].plot(x,y,color='green',ls='-',marker='o',markersize=10,markerfacecolor='yellow')
axes[5].plot(x,y,color='green',ls='-',marker='o',markersize=10,markerfacecolor='yellow',markeredgewidth=4,markeredgecolor='red')
plt.tight_layout()
```

```python
f = plt.figure()
axes1 = f.add_axes([0,0,1,1])
axes1.set_xlabel("X")
axes1.set_ylabel("Y")
axes1.set_title("y = x^2")
axes1.plot(x,y)

#zoom in
axes1.set_xlim([0,3])#scaling 
axes1.set_ylim([0,30])
```
