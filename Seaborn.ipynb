```python
#Seaborn: statistical plotting library built on top of matplotlib
import seaborn as sns, numpy as np
%matplotlib inline
```

```python
tips_data = sns.load_dataset('tips')
print(type(tips_data))
tips_data.head()
```

```python
#visualizing univariate(one variable) distribution
#histogram and KDE(gaussian kernel density estimation)
sns.distplot(tips_data['total_bill'], kde=True, bins=20, rug=True, color='red')
```

```python
sns.rugplot(tips_data['total_bill'])
```

```python
sns.kdeplot(tips_data['total_bill'])
```

```python
#scatter_plot of two variables with bivariate and univariate graphs
sns.jointplot(x='total_bill', y='tip', data=tips_data)
```

```python
#hexagonal
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind='hex')
```

```python
#regression
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind = 'reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips_data, kind = 'kde')
```

```python
#jointplot for every numerical column combination 
sns.pairplot(tips_data, hue='sex', palette='husl')
```

```python
sns.pairplot(tips_data, hue='day', palette='coolwarm')#pallete= (matplotlib colormap strings)
```

# Categorical plots

```python
tips_data.tail()
```

```python
sns.barplot(x='day',y='tip',data=tips_data,estimator=np.std)#estimator: standard deviation
sns.set_style('ticks')
sns.despine()
```

```python

sns.countplot(x='day',data=tips_data)
sns.set_style('whitegrid')
```

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='day',y='total_bill',data=tips_data,hue='sex')
```

```python
sns.set_context('poster',font_scale=2)
sns.violinplot(x='day',y='total_bill',data=tips_data,hue='sex',split=True)
```

```python
sns.set_context('notebook')
sns.stripplot(x='day',y='total_bill',data=tips_data,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips_data)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips_data)
sns.swarmplot(x='day',y='total_bill',data=tips_data,color='black')
```

```python
#general
sns.factorplot(x='day',y='total_bill',data=tips_data,kind='bar')#kind= violin,swarm,strip,bar
```

# Matrix Plots

```python
flights_data  = sns.load_dataset('flights')
flights_data.head()
```

```python
tips_data.head()
```

```python
#pairwise correlation of columns
td = tips_data.corr()
td
```

```python
sns.heatmap(td,annot=True,cmap='coolwarm')
```

```python
fp = flights_data.pivot_table(index='month',columns='year',values='passengers')
sns.heatmap(fp,linecolor='white',linewidths=3)
#seaborn is cool :)
```

```python
sns.clustermap(fp, standard_scale=1)#similar groups are close to each
```

# Grids

```python
iris = sns.load_dataset('iris')
print(iris['species'].unique())
iris.head()
```

```python
sns.pairplot(iris)
```

```python
import matplotlib.pyplot as plt

pg = sns.PairGrid(iris)
pg.map_diag(sns.distplot)
pg.map_upper(plt.scatter)
pg.map_lower(sns.kdeplot)
```

```python
fg = sns.FacetGrid(tips_data, col="time", row="smoker")
fg.map(plt.scatter, 'total_bill','tip')
```
